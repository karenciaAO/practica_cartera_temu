## 📌 3. Feature engineering sugerido

**Bandera “primer uso en Temu”:** first_use_temu = (NumeroCreditosGPrevius==0).astype(int).

**Ratios:**

- creditos_activos_ratio = NumeroCreditosGEstadoActivosPrevius/(NumeroCreditosGPrevius+1)

**Antigüedad:** meses desde FechaVinculacionCliente a FechaEvento.

**Bandera canal virtual:** de TipoMunicipioEntregaTC o CanalMunicipioEntregaTC.

**Score crediticio bucket:** 0 (sin info), bajo, medio, alto.

Estas transformaciones dan un set de features más estable y explicable.

Hagamos una auditoría de rangos y valores raros bien completa (solo prints/tablas, sin gráficos) para:

numéricas: min/max, p5/p95, conteo de negativos, ceros, IQR outliers, z-score outliers.

columnas “conteo” (créditos/pagos): chequear negativos y “no enteros”.

ratios: verificar que estén en [0,1].

categóricas: contar niveles raros (frecuencia < 0.1%) y top-10.

edad y cupo: rangos plausibles.

un resumen de “issues” por columna.

En este dataset específico, el diccionario dice explícitamente:
ScoreCrediticio <= 0 = “sin información suficiente” (no que sea “muy malo”).
Así que, para evitar introducir un sesgo que contradice la semántica del dataset, la codificación correcta por defecto es tratar <= 0 como “sin_info” (y por eso te propuse mapear negativos a 0, con un Flag_Score_Negativo por si hubiera señal residual).

```python
# ==========================================
# PIPELINE CONSOLIDADO: Limpieza + Features
# ==========================================
import numpy as np
import pandas as pd

TARGET = 'PerdidaCartera'
df = clientes.copy()  # mantén una copia de trabajo

# -------------------------
# 0) PARSEO Y FEATURES DE FECHA
# -------------------------
# FechaEvento: ISO con 'Z' (UTC) -> tz-aware -> tz-naive
df['FechaEvento_dt'] = (
    pd.to_datetime(df['FechaEvento'], errors='coerce', utc=True)
      .dt.tz_convert(None)
)

# Excel serial dates:
# - Vinculación y ÚltimoUso: base 1900 (1899-12-30)
# - PrimerUso: base 1904 (1904-01-01)
df['FechaVinculacionCliente_dt'] = pd.to_datetime(
    df['FechaVinculacionCliente'], errors='coerce', origin='1899-12-30', unit='D'
)
df['FechaUltimoUso_dt'] = pd.to_datetime(
    df['FechaUltimoUso'], errors='coerce', origin='1899-12-30', unit='D'
)
df['FechaPrimerUso_dt'] = pd.to_datetime(
    df['FechaPrimerUso'], errors='coerce', origin='1904-01-01', unit='D'
)

# Flag inconsistencia y corrección conservadora de PrimerUso
df['Flag_PrimerUsoAntesVinc'] = (
    (df['FechaPrimerUso_dt'] < df['FechaVinculacionCliente_dt']).fillna(False).astype(int)
)
mask_bad = df['FechaPrimerUso_dt'].isna() | (df['FechaPrimerUso_dt'] < df['FechaVinculacionCliente_dt'])
df['FechaPrimerUso_corr'] = np.where(mask_bad, df['FechaVinculacionCliente_dt'], df['FechaPrimerUso_dt'])

# Diferencias seguras (NaN->0, negativos->0)
def _safe_days(a, b):
    dd = (a - b).dt.days
    dd = dd.where(dd.notna(), 0)
    return np.where(dd < 0, 0, dd)

def _safe_months(a, b):
    return _safe_days(a, b) / 30.0

df['MesesDesdeVinculacion'] = _safe_months(df['FechaEvento_dt'], df['FechaVinculacionCliente_dt'])
df['MesesDesdePrimerUso']   = _safe_months(df['FechaEvento_dt'], df['FechaPrimerUso_corr'])
df['DiasDesdeUltimoUso']    = _safe_days(df['FechaEvento_dt'], df['FechaUltimoUso_dt'])

# Posible fuga (no se usa la fecha cruda, solo dejamos flag)
df['Flag_UltimoUsoPosterior'] = (df['FechaUltimoUso_dt'] > df['FechaEvento_dt']).fillna(False).astype(int)

# -------------------------
# 1) LIMPIEZA CATEGÓRICAS
# -------------------------
# Genero (27 -> 'Desconocido'), nulos a 'Desconocido'
if 'Genero' in df.columns:
    df['Genero'] = df['Genero'].replace({27: 'Desconocido'}).fillna('Desconocido')

# TipoMunicipioEntregaTC: fix encoding y nulos
if 'TipoMunicipioEntregaTC' in df.columns:
    df['TipoMunicipioEntregaTC'] = (
        df['TipoMunicipioEntregaTC']
          .replace({'PEQUEÃ‘O': 'PEQUEÑO'})
          .fillna('Desconocido')
    )

# CanalMunicipioEntregaTC: nulos a 'Desconocido'
if 'CanalMunicipioEntregaTC' in df.columns:
    df['CanalMunicipioEntregaTC'] = df['CanalMunicipioEntregaTC'].fillna('Desconocido')

# UsoAppWeb / CategoriaPrincipalCredito: nulos a 'Desconocido'
for c in ['UsoAppWeb','CategoriaPrincipalCredito']:
    if c in df.columns:
        df[c] = df[c].fillna('Desconocido')

# Agrupar categorías raras en CategoriaPrincipalCredito (<0.1%)
def _lump_rare(series, min_frac=0.001, rare_label='OtrosRare'):
    vc = series.value_counts(dropna=False)
    cutoff = series.shape[0] * min_frac
    rare_levels = vc[vc < cutoff].index
    return series.where(~series.isin(rare_levels), rare_label)

if 'CategoriaPrincipalCredito' in df.columns:
    df['CategoriaPrincipalCredito'] = _lump_rare(df['CategoriaPrincipalCredito'].astype(str), min_frac=0.001)

# -------------------------
# 2) USABILIDAD CUPO
# -------------------------
# Asegurar numérica, cap por negocio en [0,2] y mediana para NaN
df['UsabilidadCupo'] = pd.to_numeric(df.get('UsabilidadCupo', np.nan), errors='coerce')
df['Flag_Usab_NaN']     = df['UsabilidadCupo'].isna().astype(int)
df['Flag_Usab_Outlier'] = ((df['UsabilidadCupo'] < 0) | (df['UsabilidadCupo'] > 2)).astype(int)
df.loc[df['UsabilidadCupo'] < 0, 'UsabilidadCupo'] = np.nan
df.loc[df['UsabilidadCupo'] > 2, 'UsabilidadCupo'] = np.nan
df['UsabilidadCupo'] = df['UsabilidadCupo'].fillna(df['UsabilidadCupo'].median())

# -------------------------
# 3) NULOS NUMÉRICOS (imput + flags)
# -------------------------
num_cols_candidates = [
    'DiasMaximosMoraCreditosGenerados',
    'NumeroCreditosGPrevius','NumeroCreditosGCanalFPrevius','NumeroCreditosGCanalVPrevius',
    'NumeroCreditosGEstadoActivosPrevius','NumeroCreditosGEstadoPagadosPrevius',
    'NumeroCreditosLPrevius','NumeroCreditosLEstadoActivosPrevius','NumeroCreditosLEstadoPagadosPrevius',
    'TotalPagosEfectuadosGlobalmentePrevius','TotalPagosEfectuadosLocalmentePrevius',
    'NumeroIntentosFallidos','CupoAprobado','ScoreCrediticio','Edad',
    'MesesDesdeVinculacion','MesesDesdePrimerUso','DiasDesdeUltimoUso'
]
for c in [x for x in num_cols_candidates if x in df.columns]:
    df[f'Flag_{c}_NaN'] = df[c].isna().astype(int)
    if c.startswith('NumeroCreditos') or c.startswith('TotalPagos') or c in ['DiasMaximosMoraCreditosGenerados','Edad','NumeroIntentosFallidos']:
        df[c] = df[c].fillna(0)

# -------------------------
# 4) SCORE & CUPO
# -------------------------
# Score negativo -> 0 + flag. 0 se interpreta como "sin info"
if 'ScoreCrediticio' in df.columns:
    df['Flag_Score_Negativo'] = (df['ScoreCrediticio'] < 0).astype(int)
    df.loc[df['ScoreCrediticio'] < 0, 'ScoreCrediticio'] = 0
    df['ScoreSinInfo'] = (df['ScoreCrediticio'] == 0).astype(int)
    if df['ScoreCrediticio'].isna().any():
        med_pos = df.loc[df['ScoreCrediticio'] > 0, 'ScoreCrediticio'].median()
        df['ScoreCrediticio'] = df['ScoreCrediticio'].fillna(med_pos)

# CupoAprobado: versión log + imput mediana si hiciera falta
if 'CupoAprobado' in df.columns:
    df['log_CupoAprobado'] = np.log1p(df['CupoAprobado'])
    if df['log_CupoAprobado'].isna().any():
        df['log_CupoAprobado'] = df['log_CupoAprobado'].fillna(df['log_CupoAprobado'].median())

# -------------------------
# 5) FEATURES DERIVADAS
# -------------------------
# Bandera “primer uso en Temu”
if 'NumeroCreditosGPrevius' in df.columns and 'Flag_PrimerUsoTemu' not in df.columns:
    df['Flag_PrimerUsoTemu'] = (df['NumeroCreditosGPrevius'] == 0).astype(int)

# Ratios (clip [0,1])
if {'TotalPagosEfectuadosLocalmentePrevius','TotalPagosEfectuadosGlobalmentePrevius'}.issubset(df.columns):
    df['ratio_pagos_local_global'] = (
        df['TotalPagosEfectuadosLocalmentePrevius'].fillna(0) /
        (df['TotalPagosEfectuadosGlobalmentePrevius'].fillna(0) + 1.0)
    ).replace([np.inf,-np.inf], np.nan).fillna(0).clip(0,1)

if {'NumeroCreditosGEstadoActivosPrevius','NumeroCreditosGPrevius'}.issubset(df.columns):
    df['creditos_activos_ratio'] = (
        df['NumeroCreditosGEstadoActivosPrevius'].fillna(0) /
        (df['NumeroCreditosGPrevius'].fillna(0) + 1.0)
    ).replace([np.inf,-np.inf], np.nan).fillna(0).clip(0,1)

# Bandera canal virtual (si cualquiera lo indica)
if {'CanalMunicipioEntregaTC','TipoMunicipioEntregaTC'}.issubset(df.columns):
    df['Flag_CanalVirtual'] = (
        (df['CanalMunicipioEntregaTC'].astype(str).str.lower() == 'virtual') |
        (df['TipoMunicipioEntregaTC'].astype(str).str.upper() == 'VIRTUAL')
    ).astype(int)

# Buckets de ScoreCrediticio (explicable) para >0
if 'ScoreCrediticio' in df.columns:
    df['ScoreBucket'] = 'sin_info'
    mask_pos = df['ScoreCrediticio'] > 0
    if mask_pos.sum() > 0:
        q33, q66 = df.loc[mask_pos, 'ScoreCrediticio'].quantile([0.33, 0.66]).values
        df.loc[mask_pos & (df['ScoreCrediticio'] <= q33), 'ScoreBucket'] = 'bajo'
        df.loc[mask_pos & (df['ScoreCrediticio'] >  q33) & (df['ScoreCrediticio'] <= q66), 'ScoreBucket'] = 'medio'
        df.loc[mask_pos & (df['ScoreCrediticio'] >  q66), 'ScoreBucket'] = 'alto'
    df['ScoreBucket'] = pd.Categorical(df['ScoreBucket'], categories=['sin_info','bajo','medio','alto'], ordered=True)

# Edad: rango plausible + flag
if 'Edad' in df.columns:
    df['Flag_Edad_Out'] = (~df['Edad'].between(18, 100, inclusive='both')).fillna(False).astype(int)
    df.loc[df['Flag_Edad_Out'] == 1, 'Edad'] = np.nan
    df['Edad'] = df['Edad'].fillna(df['Edad'].median())

# -------------------------
# 6) CAPEO (winsorize) p99 + flags
# -------------------------
cap_cols = [
    'DiasDesdeUltimoUso','MesesDesdeVinculacion',
    'TotalPagosEfectuadosGlobalmentePrevius','TotalPagosEfectuadosLocalmentePrevius',
    'NumeroCreditosGPrevius','NumeroCreditosGCanalFPrevius','NumeroCreditosGCanalVPrevius',
    'NumeroCreditosGEstadoActivosPrevius','NumeroCreditosGEstadoPagadosPrevius',
    'NumeroCreditosLPrevius','NumeroCreditosLEstadoActivosPrevius','NumeroCreditosLEstadoPagadosPrevius'
]
for c in [x for x in cap_cols if x in df.columns]:
    p99 = df[c].quantile(0.99)
    df[f'Flag_{c}_Capped'] = (df[c] > p99).astype(int)
    df[c] = np.where(df[c] > p99, p99, df[c])

# -------------------------
# 7) LISTA FINAL DE FEATURES
# -------------------------
drop_from_features = [
    'IdentificadorCliente','FechaEvento','FechaVinculacionCliente','FechaPrimerUso','FechaUltimoUso',
    'FechaEvento_dt','FechaVinculacionCliente_dt','FechaPrimerUso_dt','FechaUltimoUso_dt','FechaPrimerUso_corr',
    'CodigoAlmacenEntregaTC','CodigoAlmacenEntregaTC_str','AlmacenTop20',
    'CodigoMunicipioEntregaTC','MunicipioCat','MunicipioTop20','MesCompra',
    'DiasMora'  # fuga
]

num_final = [
    'UsabilidadCupo','MesesDesdeVinculacion','MesesDesdePrimerUso','DiasDesdeUltimoUso',
    'NumeroCreditosGPrevius','NumeroCreditosGCanalFPrevius','NumeroCreditosGCanalVPrevius',
    'NumeroCreditosGEstadoActivosPrevius','NumeroCreditosGEstadoPagadosPrevius',
    'NumeroCreditosLPrevius','NumeroCreditosLEstadoActivosPrevius','NumeroCreditosLEstadoPagadosPrevius',
    'TotalPagosEfectuadosGlobalmentePrevius','TotalPagosEfectuadosLocalmentePrevius',
    'NumeroIntentosFallidos','ScoreCrediticio','ScoreSinInfo','CupoAprobado','log_CupoAprobado','Edad',
    'Flag_Usab_NaN','Flag_Usab_Outlier','Flag_PrimerUsoAntesVinc','Flag_UltimoUsoPosterior','Flag_Edad_Out',
    'ratio_pagos_local_global','creditos_activos_ratio','Flag_Score_Negativo'
] + [f'Flag_{c}_NaN' for c in num_cols_candidates if f'Flag_{c}_NaN' in df.columns] \
  + [f'Flag_{c}_Capped' for c in cap_cols if f'Flag_{c}_Capped' in df.columns]

cat_final = [
    'Genero','TipoMunicipioEntregaTC','CanalMunicipioEntregaTC','UsoAppWeb','CategoriaPrincipalCredito',
    'Flag_PrimerUsoTemu','ScoreBucket'
]

# Filtrado final (existencia y exclusiones)
num_final = [c for c in num_final if c in df.columns and c not in drop_from_features]
cat_final = [c for c in cat_final if c in df.columns and c not in drop_from_features]

# -------------------------
# 8) CONSTRUCCIÓN X / y
# -------------------------
X = df[num_final + cat_final].copy()
y = df[TARGET].copy()

print("\n=== Resumen columnas finales ===")
print("Numéricas (", len(num_final), "):", num_final[:15], "...")
print("Categóricas (", len(cat_final), "):", cat_final)
print("X shape:", X.shape, "| y rate (1):", round(y.mean(), 3))

# -------------------------
# 9) AUDITORÍAS RÁPIDAS
# -------------------------
# Correlación numéricas con el target
try:
    print("\n=== Correlación numéricas con", TARGET, "===")
    print(X[num_final].corrwith(y).sort_values(ascending=False).to_string())
except Exception as e:
    print("Aviso correlaciones:", e)

# Crosstab para categóricas con <=40 niveles
for c in cat_final:
    if X[c].nunique() <= 40:
        print(f"\n=== {c} vs {TARGET} ===")
        print(pd.crosstab(X[c], y, normalize='index').round(3).to_string())

# Chequeos de conteos y ratios
count_like = [
    'NumeroCreditosGPrevius','NumeroCreditosGCanalFPrevius','NumeroCreditosGCanalVPrevius',
    'NumeroCreditosGEstadoActivosPrevius','NumeroCreditosGEstadoPagadosPrevius',
    'NumeroCreditosLPrevius','NumeroCreditosLEstadoActivosPrevius','NumeroCreditosLEstadoPagadosPrevius',
    'TotalPagosEfectuadosGlobalmentePrevius','TotalPagosEfectuadosLocalmentePrevius',
    'NumeroIntentosFallidos'
]
ratio_like = ['UsabilidadCupo','ratio_pagos_local_global','creditos_activos_ratio']

print("\n=== Chequeo ratios [0,1] ===")
for c in [x for x in ratio_like if x in df.columns]:
    s = pd.to_numeric(df[c], errors='coerce')
    print(f"{c}: min={s.min():.4f} max={s.max():.4f}")

print("\n=== Chequeo conteos (>=0 y enteros) ===")
rows = []
for c in [x for x in count_like if x in df.columns]:
    s = pd.to_numeric(df[c], errors='coerce')
    rows.append([c, int((s<0).sum()), int(((s.dropna()%1)!=0).sum()), int(s.max())])
if rows:
    print(pd.DataFrame(rows, columns=['col','negativos','no_enteros','max']).to_string(index=False))

# -------------------------
# 10) SALIDA
# -------------------------
# df queda limpio con todas las features/flags
clientes = df
```

```python
# ============================
# BASELINE con split temporal
# ============================
import numpy as np
import pandas as pd

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import (
    roc_auc_score, average_precision_score, brier_score_loss,
    classification_report, confusion_matrix, precision_recall_curve
)

# 0) DF limpio y listas finales
df = clientes
X = df[num_final + cat_final].copy()
y = df['PerdidaCartera'].astype(int).copy()

# 1) Split temporal (80/20 por fecha)
cutoff = df['FechaEvento_dt'].quantile(0.8)
train_idx = df['FechaEvento_dt'] <= cutoff
valid_idx = df['FechaEvento_dt'] >  cutoff

X_tr, X_va = X.loc[train_idx].copy(), X.loc[valid_idx].copy()
y_tr, y_va = y.loc[train_idx], y.loc[valid_idx]

print(f"Split temporal: train={X_tr.shape}, valid={X_va.shape}, corte={cutoff.date()}")

# 2) Poda de columnas constantes/casi-constantes (solo usando TRAIN)
low_var_cols = [c for c in X_tr.columns if X_tr[c].nunique(dropna=False) <= 1]
near_const = []
for c in X_tr.columns:
    vc = X_tr[c].value_counts(dropna=False, normalize=True)
    if len(vc) > 0 and vc.iloc[0] >= 0.999:
        near_const.append(c)

cols_to_drop = sorted(set(low_var_cols + near_const))
if cols_to_drop:
    print("Columnas podadas (baja varianza):", cols_to_drop)
    X_tr.drop(columns=cols_to_drop, inplace=True)
    X_va.drop(columns=cols_to_drop, inplace=True, errors='ignore')

# 3) Definir columnas num/cat efectivas tras la poda
cat_cols = [c for c in cat_final if c in X_tr.columns]
num_cols = [c for c in X_tr.columns if c not in cat_cols]

# 4) Preprocesamiento con IMPUTACIÓN (evita NaN) + OHE
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore", sparse=False))
])

pre = ColumnTransformer(
    transformers=[
        ("cat", categorical_transformer, cat_cols),
        ("num", numeric_transformer, num_cols),
    ],
    remainder="drop"
)

# Helper evaluación
def evaluate_proba(y_true, y_proba, name="model"):
    auc = roc_auc_score(y_true, y_proba)
    ap  = average_precision_score(y_true, y_proba)
    br  = brier_score_loss(y_true, y_proba)
    print(f"[{name}] ROC-AUC={auc:.3f} | PR-AUC={ap:.3f} | Brier={br:.3f}")
    p, r, thr = precision_recall_curve(y_true, y_proba)
    f1 = 2*p*r/(p+r+1e-12)
    best = int(np.argmax(f1))
    best_thr = float(thr[best-1]) if best > 0 and best-1 < len(thr) else 0.5
    print(f"[{name}] best-F1={f1[best]:.3f} @ thr={best_thr:.3f}")
    y_hat = (y_proba >= best_thr).astype(int)
    print(f"[{name}] ConfMatrix @thr={best_thr:.3f}:\n", confusion_matrix(y_true, y_hat))
    print(f"[{name}] Report @thr={best_thr:.3f}:\n", classification_report(y_true, y_hat, digits=3))
    return dict(roc_auc=auc, pr_auc=ap, brier=br, thr=best_thr)

# 5A) Regresión logística + calibración
logit = Pipeline(steps=[
    ("pre", pre),
    ("clf", LogisticRegression(
        solver="saga", penalty="l2",
        class_weight="balanced",
        max_iter=1000, n_jobs=-1, random_state=42
    ))
])

logit.fit(X_tr, y_tr)

cal_logit = CalibratedClassifierCV(logit, method="sigmoid", cv="prefit")
cal_logit.fit(X_va, y_va)
proba_logit = cal_logit.predict_proba(X_va)[:, 1]
metrics_logit = evaluate_proba(y_va, proba_logit, name="Logit+Cal")

# Nombres de features transformadas (coefs de la LR antes de calibrar)
fitted_pre = logit.named_steps["pre"]
feat_names = fitted_pre.get_feature_names_out()  # usa nombres tipo 'cat__Genero_Femenino' y 'num__Edad'
coef = logit.named_steps["clf"].coef_[0]
topk = np.argsort(np.abs(coef))[::-1][:20]
print("\n[Logit] Top 20 coef (abs):")
for i in topk:
    print(f"{feat_names[i]:45s} {coef[i]: .4f}")

# 5B) HistGradientBoosting + calibración
# (requiere entrada densa → ya usamos OHE con sparse=False)
hgb = Pipeline(steps=[
    ("pre", pre),
    ("clf", HistGradientBoostingClassifier(
        learning_rate=0.07, max_depth=None, max_leaf_nodes=31,
        l2_regularization=1.0, random_state=42
    ))
])

hgb.fit(X_tr, y_tr)

cal_hgb = CalibratedClassifierCV(hgb, method="isotonic", cv="prefit")
cal_hgb.fit(X_va, y_va)
proba_hgb = cal_hgb.predict_proba(X_va)[:, 1]
metrics_hgb = evaluate_proba(y_va, proba_hgb, name="HGB+Cal")

# Importancias (del HGB antes de calibrar)
hgb_fitted = hgb.named_steps["clf"]
imp = getattr(hgb_fitted, "feature_importances_", None)
if imp is not None:
    names = fitted_pre.get_feature_names_out()
    top = np.argsort(imp)[::-1][:20]
    print("\n[HGB] Top 20 importancias:")
    for i in top:
        print(f"{names[i]:45s} {imp[i]: .5f}")
else:
    print("\n[HGB] Este estimador no expone feature_importances_ en tu versión.")

# 6) Umbral recomendado (por F1 en valid del HGB)
best_thr = metrics_hgb["thr"]
print(f"\nUmbral recomendado (HGB, valid/F1): {best_thr:.3f}")
```

```python
# ===============================================
# Post-análisis pro: coeficientes, explicabilidad,
# calibración, deciles, backtest por mes, umbral por costo
# ===============================================
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
from sklearn.metrics import (
    roc_auc_score, average_precision_score, brier_score_loss,
    precision_recall_curve
)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.calibration import CalibratedClassifierCV

warnings.filterwarnings("ignore")

# -------- Helpers generales --------
def get_calibrated_base(cal):
    """
    Devuelve el estimador base del CalibratedClassifierCV de forma robusta
    para versiones distintas de sklearn.
    """
    try:
        return cal.calibrated_classifiers_[0].estimator  # sklearn >= 1.2
    except Exception:
        try:
            return cal.base_estimator  # fallback
        except Exception:
            return None

def proba_to_deciles(y_true, y_proba, n=10):
    dfp = pd.DataFrame({'y': y_true, 'p': y_proba}).sort_values('p', ascending=False)
    dfp['decile'] = pd.qcut(dfp['p'].rank(method='first'), q=n, labels=list(range(1, n+1)))
    out = dfp.groupby('decile').agg(
        n=('y','size'),
        eventos=('y','sum'),
        tasa=('y','mean'),
        p_min=('p','min'),
        p_max=('p','max')
    ).reset_index()
    out['acum_eventos'] = out['eventos'].cumsum()
    out['lift'] = out['tasa'] / dfp['y'].mean()
    return out

def calib_curve(y_true, y_proba, bins=10):
    dfc = pd.DataFrame({'y':y_true, 'p':y_proba})
    dfc['bin'] = pd.qcut(dfc['p'], q=bins, duplicates='drop')
    g = dfc.groupby('bin').agg(
        p_mean=('p','mean'),
        y_rate=('y','mean'),
        n=('y','size')
    ).reset_index()
    return g

def choose_threshold_by_cost(y_true, y_proba, c_fp=1.0, c_fn=5.0):
    p, r, thr = precision_recall_curve(y_true, y_proba)
    # reconstruimos precisión/recall en thresholds y calculamos costos
    thr = np.append(thr, 1.0)  # asegurar última
    costs = []
    y = np.asarray(y_true)
    for t in thr:
        yhat = (y_proba >= t).astype(int)
        fp = np.sum((y==0) & (yhat==1))
        fn = np.sum((y==1) & (yhat==0))
        cost = c_fp*fp + c_fn*fn
        costs.append((t, cost))
    best_t, best_cost = min(costs, key=lambda x: x[1])
    return best_t, best_cost

# ==================================================
# 0) Garantizar objetos necesarios (si no existen)
# ==================================================
TARGET = 'PerdidaCartera'

# Si no existen, intenta reconstruir mínimos.
if 'X_tr' not in globals() or 'X_va' not in globals():
    # asumimos que ya definiste num_final/cat_final, pre, etc. en el entrenamiento anterior
    df = clientes.copy()
    X = df[num_final + cat_final].copy()
    y = df[TARGET].astype(int).copy()

    # Poda
    low_var_cols = [c for c in X.columns if X[c].nunique(dropna=False) <= 1]
    X.drop(columns=low_var_cols, inplace=True)

    cat_cols = [c for c in cat_final if c in X.columns]
    num_cols = [c for c in X.columns if c not in cat_cols]

    # Split temporal
    cutoff = df.loc[X.index, 'FechaEvento_dt'].quantile(0.8)
    train_idx = df['FechaEvento_dt'] <= cutoff
    valid_idx = df['FechaEvento_dt'] >  cutoff

    X_tr, X_va = X.loc[train_idx], X.loc[valid_idx]
    y_tr, y_va = y.loc[train_idx], y.loc[valid_idx]

    # Preprocesador
    ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    num_imputer = SimpleImputer(strategy="median")
    pre = ColumnTransformer(
        transformers=[
            ('cat', ohe, cat_cols),
            ('num', num_imputer, num_cols)
        ],
        remainder='drop'
    )

    # Modelos (rápidos)
    logit = Pipeline(steps=[
        ('pre', pre),
        ('clf', LogisticRegression(
            solver='saga', penalty='elasticnet', l1_ratio=0.5,
            C=1.0, class_weight='balanced', max_iter=5000,
            n_jobs=-1, random_state=42
        ))
    ])
    logit.fit(X_tr, y_tr)
    cal_logit = CalibratedClassifierCV(logit, method='sigmoid', cv='prefit').fit(X_va, y_va)

    hgb = Pipeline(steps=[
        ('pre', pre),
        ('clf', HistGradientBoostingClassifier(
            learning_rate=0.07, max_depth=None, max_leaf_nodes=31,
            l2_regularization=1.0, random_state=42
        ))
    ])
    hgb.fit(X_tr, y_tr)
    cal_hgb = CalibratedClassifierCV(hgb, method='isotonic', cv='prefit').fit(X_va, y_va)

# ==================================================
# 1) Coeficientes del Logit (fix acceso CalibratedCV)
# ==================================================
try:
    base_logit = get_calibrated_base(cal_logit)  # Pipeline(pre, clf)
    assert base_logit is not None, "No se pudo acceder al estimador base del logit calibrado."
    pre_fitted = base_logit.named_steps['pre']
    clf_logit  = base_logit.named_steps['clf']
    try:
        feat_names = pre_fitted.get_feature_names_out()
    except Exception:
        cat_names = pre_fitted.named_transformers_['cat'].get_feature_names_out(cat_cols)
        feat_names = np.concatenate([cat_names, np.array(num_cols, dtype=object)])
    coefs = clf_logit.coef_[0]
    top_idx = np.argsort(np.abs(coefs))[::-1][:20]
    print("\n[Logit] Top 20 coef (abs) — FIXED:")
    for i in top_idx:
        print(f"{feat_names[i]:40s} {coefs[i]: .4f}")
except Exception as e:
    print("\n[Logit] Aviso: no fue posible listar coeficientes (", str(e), ")")

# ==================================================
# 2) Importancias para HGB por Permutation Importance
#    (funciona incluso si no hay feature_importances_)
# ==================================================
try:
    # Usamos el pipeline calibrado completo para que permute sobre columnas originales
    r = permutation_importance(
        estimator=cal_hgb, X=X_va, y=y_va,
        n_repeats=3, random_state=42, scoring='roc_auc'
    )
    pi = pd.DataFrame({
        'feature': X_va.columns,
        'importance_mean': r.importances_mean,
        'importance_std': r.importances_std
    }).sort_values('importance_mean', ascending=False)
    print("\n[HGB] Top 20 permutation importance (val):")
    print(pi.head(20).to_string(index=False))
except Exception as e:
    print("\n[HGB] Aviso: no fue posible calcular permutation importance (", str(e), ")")

# ==================================================
# 3) Curva de calibración / confiabilidad
# ==================================================
proba_va_hgb = cal_hgb.predict_proba(X_va)[:,1]
cal_df = calib_curve(y_va, proba_va_hgb, bins=10)
print("\n[Calibración] Promedio por decil de probas (val):")
print(cal_df.to_string(index=False))

plt.figure()
plt.plot([0,1],[0,1], linestyle='--')
plt.plot(cal_df['p_mean'], cal_df['y_rate'], marker='o')
plt.xlabel('Probabilidad promedio (bin)')
plt.ylabel('Tasa real (bin)')
plt.title('Reliability curve — HGB calibrado (valid)')
plt.show()

# ==================================================
# 4) Tabla de deciles / lift (valid)
# ==================================================
dec = proba_to_deciles(y_va, proba_va_hgb, n=10)
print("\n[Ranking] Tabla de deciles (valid):")
print(dec.to_string(index=False))

# ==================================================
# 5) Backtest por MES en el split de validación
# ==================================================
val_index = X_va.index
val_month = pd.to_datetime(df.loc[val_index, 'FechaEvento_dt']).dt.to_period('M')
bt = []
for m in sorted(val_month.unique()):
    mask = (val_month == m)
    if mask.sum() < 100:  # evita meses con muy pocas obs
        continue
    auc_m = roc_auc_score(y_va[mask], proba_va_hgb[mask])
    pr_m  = average_precision_score(y_va[mask], proba_va_hgb[mask])
    rate  = y_va[mask].mean()
    bt.append([str(m), mask.sum(), rate, auc_m, pr_m])
bt_df = pd.DataFrame(bt, columns=['mes','n','tasa_positiva','roc_auc','pr_auc'])
print("\n[Backtest] Métricas por mes (valid):")
print(bt_df.to_string(index=False))

# ==================================================
# 6) Umbral por costo (ejemplo: FN cuesta 5x FP)
# ==================================================
thr_cost, cost_val = choose_threshold_by_cost(y_va, proba_va_hgb, c_fp=1.0, c_fn=5.0)
print(f"\n[Threshold por costo] Mejor thr≈{thr_cost:.3f} (costo esperado={cost_val:,.0f} con c_fp=1, c_fn=5)")
```

```python
import numpy as np
import pandas as pd
import warnings

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import (
    roc_auc_score, average_precision_score, brier_score_loss,
    confusion_matrix, classification_report, precision_recall_curve
)
from sklearn.inspection import permutation_importance

warnings.filterwarnings("ignore")

# =============================
# 0) Datos y poda de constantes
# =============================
TARGET = 'PerdidaCartera'
df = clientes.copy()

X = df[num_final + cat_final].copy()
y = df[TARGET].astype(int).copy()

# Poda columnas casi-constantes (evita fugas de info inútil)
low_var_cols = [c for c in X.columns if X[c].nunique(dropna=False) <= 1]
if low_var_cols:
    X.drop(columns=low_var_cols, inplace=True)
    print("Poda de columnas casi-constantes:", len(low_var_cols), "->", low_var_cols)

# Categóricas y numéricas finales (sobre X ya podado)
cat_cols = [c for c in cat_final if c in X.columns]
num_cols = [c for c in X.columns if c not in cat_cols]

# ==================================
# 1) Split temporal 80/20 por fecha
# ==================================
cutoff = df.loc[X.index, 'FechaEvento_dt'].quantile(0.8)
train_idx = df['FechaEvento_dt'] <= cutoff
valid_idx = df['FechaEvento_dt'] >  cutoff

X_tr, X_va = X.loc[train_idx], X.loc[valid_idx]
y_tr, y_va = y.loc[train_idx], y.loc[valid_idx]

print(f"Split temporal -> train={X_tr.shape}, valid={X_va.shape}, corte={cutoff.date()} (≈80/20)")

# ==========================================
# 2) Preprocesamiento (claro y consistente)
#    - Num: impute (median) + StandardScaler
#    - Cat: impute (most_frequent) + OHE
# ==========================================
pre = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imp', SimpleImputer(strategy='median')),
            ('sc', StandardScaler())
        ]), num_cols),
        ('cat', Pipeline(steps=[
            ('imp', SimpleImputer(strategy='most_frequent')),
            ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
        ]), cat_cols)
    ],
    remainder='drop'
)

# ==========================================
# 3) Modelos
# ==========================================
# 3A) Logit interpretable (elastic net balanceado)
logit = Pipeline(steps=[
    ('pre', pre),
    ('clf', LogisticRegression(
        solver='saga', penalty='elasticnet', l1_ratio=0.5,
        C=1.0, class_weight='balanced', max_iter=5000,
        random_state=42, n_jobs=-1
    ))
])
logit.fit(X_tr, y_tr)

# Calibración externa (sigmoid para logit)
cal_logit = CalibratedClassifierCV(logit, method='sigmoid', cv='prefit')
cal_logit.fit(X_va, y_va)

# 3B) HGB (no lineal, fuerte)
hgb = Pipeline(steps=[
    ('pre', pre),
    ('clf', HistGradientBoostingClassifier(
        learning_rate=0.07, max_depth=None, max_leaf_nodes=31,
        l2_regularization=1.0, random_state=42
    ))
])
hgb.fit(X_tr, y_tr)

# Calibración externa (isotónica va muy bien con árboles)
cal_hgb = CalibratedClassifierCV(hgb, method='isotonic', cv='prefit')
cal_hgb.fit(X_va, y_va)

# ==========================================
# 4) Evaluación utilitaria y umbral por F1
# ==========================================
def evaluate_proba(y_true, y_proba, name="model"):
    auc = roc_auc_score(y_true, y_proba)
    ap  = average_precision_score(y_true, y_proba)
    br  = brier_score_loss(y_true, y_proba)
    print(f"[{name}] ROC-AUC={auc:.3f} | PR-AUC={ap:.3f} | Brier={br:.3f}")
    p, r, thr = precision_recall_curve(y_true, y_proba)
    f1 = 2*p*r/(p+r+1e-12)
    best = np.argmax(f1)
    best_thr = (thr[best-1] if best>0 and best-1 < len(thr) else 0.5)
    print(f"[{name}] best-F1={f1[best]:.3f} @ thr={best_thr:.3f}")
    y_hat = (y_proba >= best_thr).astype(int)
    print(f"[{name}] ConfMatrix @thr={best_thr:.3f}:\n", confusion_matrix(y_true, y_hat))
    print(f"[{name}] Report @thr={best_thr:.3f}:\n", classification_report(y_true, y_hat, digits=3))
    return dict(roc_auc=auc, pr_auc=ap, brier=br, thr=best_thr)

proba_logit = cal_logit.predict_proba(X_va)[:,1]
metrics_logit = evaluate_proba(y_va, proba_logit, name="Logit+Cal")

proba_hgb = cal_hgb.predict_proba(X_va)[:,1]
metrics_hgb = evaluate_proba(y_va, proba_hgb, name="HGB+Cal")

# ==========================================
# 5) Coeficientes del Logit — bien impresos
#    - Para num: odds ratio por +1 std (porque escalamos)
#    - Para cat: coef de la dummy (respecto a la base)
# ==========================================
# Recuperar nombres de features transformadas
pre_fitted = logit.named_steps['pre']
num_names = np.array(num_cols)
cat_ohe = pre_fitted.named_transformers_['cat'].named_steps['ohe']
cat_names = cat_ohe.get_feature_names_out(cat_cols)
feat_names = np.r_[num_names, cat_names]

coef = logit.named_steps['clf'].coef_[0]  # ya entrenado sobre estandarizado

# Construimos tabla legible
rows = []
for fname, c in zip(feat_names, coef):
    if fname in num_cols:
        # para num, el scaler es std=1 => OR por +1 std = exp(coef)
        or_ = np.exp(c)
        rows.append([fname, c, or_, 'num (+1 std)'])
    else:
        rows.append([fname, c, np.exp(c), 'cat (dummy)'])

coef_df = pd.DataFrame(rows, columns=['feature','coef','odds_ratio','tipo']) \
          .sort_values(by='coef', key=lambda s: np.abs(s), ascending=False)

print("\n[Logit] Top 20 coef (abs) — coef y odds_ratio (sentido claro):")
with pd.option_context('display.float_format', '{:,.6f}'.format):
    print(coef_df.head(20).to_string(index=False))

# ==========================================
# 6) HGB — Permutation Importance (val)
# ==========================================
try:
    r = permutation_importance(
        estimator=cal_hgb, X=X_va, y=y_va,
        n_repeats=3, random_state=42, scoring='roc_auc'
    )
    pi = pd.DataFrame({
        'feature': X_va.columns,
        'importance_mean': r.importances_mean,
        'importance_std': r.importances_std
    }).sort_values('importance_mean', ascending=False)
    print("\n[HGB] Top 20 permutation importance (val):")
    with pd.option_context('display.float_format', '{:,.6f}'.format):
        print(pi.head(20).to_string(index=False))
except Exception as e:
    print("\n[HGB] Aviso: no fue posible calcular permutation importance (", str(e), ")")

# ==========================================
# 7) Nota corta de lectura
# ==========================================
print("\nNOTA:")
print("- Si antes veías 0.0000 era por redondeo y falta de estandarización.")
print("- Ahora, para numéricas: interpreta el 'odds_ratio' como el cambio en odds por +1 desviación estándar.")
print("- Para categóricas: el coeficiente (y su exp) es contra la categoría base generada por OHE.")
```

```python
# =========================================================
# PRUEBA TÉCNICA: Validación temporal + Holdout + Predicción
# (simple, clara, sin fugas; con df_nuevo "futuro real")
# =========================================================
import numpy as np
import pandas as pd
import warnings, joblib, os

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import (
    roc_auc_score, average_precision_score, brier_score_loss,
    precision_recall_curve, confusion_matrix, classification_report
)

warnings.filterwarnings("ignore")

# -----------------------
# 0) Datos base
# -----------------------
TARGET = 'PerdidaCartera'
assert 'FechaEvento_dt' in clientes.columns, "Falta FechaEvento_dt en el DataFrame."

# Matriz final (ya limpia por tus pasos anteriores)
X_full = clientes[num_final + cat_final].copy()
y_full = clientes[TARGET].astype(int).copy()
dates  = clientes['FechaEvento_dt'].copy()

# Poda columnas casi-constantes (por seguridad y velocidad)
low_var_cols = [c for c in X_full.columns if X_full[c].nunique(dropna=False) <= 1]
if low_var_cols:
    X_full.drop(columns=low_var_cols, inplace=True)
    print("Poda de columnas casi-constantes:", low_var_cols)

# Columnas numéricas y categóricas (post-poda)
cat_cols = [c for c in cat_final if c in X_full.columns]
num_cols = [c for c in X_full.columns if c not in cat_cols]

# -----------------------
# 1) Preprocesamiento
# -----------------------
pre = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imp', SimpleImputer(strategy='median')),
            ('sc', StandardScaler())
        ]), num_cols),
        ('cat', Pipeline(steps=[
            ('imp', SimpleImputer(strategy='most_frequent')),
            ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
        ]), cat_cols)
    ],
    remainder='drop'
)

# -----------------------
# 2) Modelos candidatos
# -----------------------
logit = Pipeline(steps=[
    ('pre', pre),
    ('clf', LogisticRegression(
        solver='saga', penalty='elasticnet', l1_ratio=0.5,
        C=1.0, class_weight='balanced', max_iter=5000,
        random_state=42, n_jobs=-1
    ))
])

hgb = Pipeline(steps=[
    ('pre', pre),
    ('clf', HistGradientBoostingClassifier(
        learning_rate=0.07, max_depth=None, max_leaf_nodes=31,
        l2_regularization=1.0, random_state=42
    ))
])

candidates = {
    'Logit': dict(pipe=logit, calib='sigmoid'),
    'HGB'  : dict(pipe=hgb,   calib='isotonic'),
}

# -------------------------------------------------------
# 3) Walk-forward CV simple (3 cortes temporales)
# -------------------------------------------------------
def rolling_eval(X, y, dates, model_dict, n_splits=3):
    qs = np.linspace(0.6, 0.9, n_splits)  # 60-90% de la línea de tiempo
    rows = []
    for q in qs:
        cutoff = dates.quantile(q)
        tr = dates < cutoff
        va = dates >= cutoff
        if tr.sum()==0 or va.sum()==0:
            continue
        pipe = model_dict['pipe']
        calib_method = model_dict['calib']

        pipe.fit(X.loc[tr], y.loc[tr])
        cal = CalibratedClassifierCV(pipe, method=calib_method, cv='prefit')
        cal.fit(X.loc[va], y.loc[va])
        proba = cal.predict_proba(X.loc[va])[:,1]

        rows.append(dict(
            cutoff=cutoff,
            roc_auc=roc_auc_score(y.loc[va], proba),
            pr_auc=average_precision_score(y.loc[va], proba),
            brier=brier_score_loss(y.loc[va], proba)
        ))
    return pd.DataFrame(rows)

cv_results = {}
for name, cfg in candidates.items():
    cv_results[name] = rolling_eval(X_full, y_full, dates, cfg, n_splits=3)
    print(f"\n{name} — Walk-forward CV:")
    print(cv_results[name].to_string(index=False))
    print(f"{name} — medias: ROC-AUC={cv_results[name]['roc_auc'].mean():.3f} | "
          f"PR-AUC={cv_results[name]['pr_auc'].mean():.3f} | "
          f"Brier={cv_results[name]['brier'].mean():.3f}")

best_name = max(cv_results.keys(), key=lambda k: cv_results[k]['pr_auc'].mean())
print(f"\n>>> Modelo elegido por PR-AUC medio: {best_name}")
best_cfg = candidates[best_name]

# -------------------------------------------------------
# 4) Holdout temporal (80/20) — examen final
# -------------------------------------------------------
holdout_cut = dates.quantile(0.8)
train_mask  = dates <= holdout_cut
holdout_mask= dates >  holdout_cut

X_tr, y_tr  = X_full.loc[train_mask],  y_full.loc[train_mask]
X_ho, y_ho  = X_full.loc[holdout_mask], y_full.loc[holdout_mask]
print(f"\nHoldout: train={X_tr.shape}, holdout={X_ho.shape}, corte={holdout_cut.date()}")

pipe = best_cfg['pipe']
calib_method = best_cfg['calib']
pipe.fit(X_tr, y_tr)

# Calibramos con holdout (también podrías reservar una parte de train para calibración)
cal = CalibratedClassifierCV(pipe, method=calib_method, cv='prefit')
cal.fit(X_ho, y_ho)

proba_ho = cal.predict_proba(X_ho)[:,1]
auc_ho = roc_auc_score(y_ho, proba_ho)
pr_ho  = average_precision_score(y_ho, proba_ho)
br_ho  = brier_score_loss(y_ho, proba_ho)
p, r, thr = precision_recall_curve(y_ho, proba_ho)
f1 = 2*p*r/(p+r+1e-12)
best = np.argmax(f1)
thr_best = (thr[best-1] if best>0 and best-1 < len(thr) else 0.5)

print(f"[HOLDOUT-{best_name}] ROC-AUC={auc_ho:.3f} | PR-AUC={pr_ho:.3f} | Brier={br_ho:.3f}")
print(f"[HOLDOUT-{best_name}] best-F1={f1[best]:.3f} @ thr={thr_best:.3f}")
yhat_ho = (proba_ho >= thr_best).astype(int)
print(f"[HOLDOUT-{best_name}] ConfMatrix @thr={thr_best:.3f}:\n", confusion_matrix(y_ho, yhat_ho))
print(f"[HOLDOUT-{best_name}] Report @thr={thr_best:.3f}:\n", classification_report(y_ho, yhat_ho, digits=3))

# -------------------------------------------------------
# 5) Guardar artefacto para producción
# -------------------------------------------------------
ART_DIR = "./artifacts_modelo"
os.makedirs(ART_DIR, exist_ok=True)

artifact = {
    "model_name": best_name,
    "pipeline": pipe,          # pre + modelo
    "calibrator": cal,         # calibrador 'prefit'
    "num_cols": num_cols,
    "cat_cols": cat_cols,
    "all_cols": list(X_full.columns),   # columnas esperadas (post-poda)
    "threshold": float(thr_best),
    "cutoff_date": pd.Timestamp(holdout_cut)
}
joblib.dump(artifact, os.path.join(ART_DIR, "modelo_calibrado.joblib"))
print(f"\nArtefactos guardados en {ART_DIR}/modelo_calibrado.joblib")

# -------------------------------------------------------
# 6) Función para predecir NUEVOS LOTES (df_nuevo)
#    - Rechaza filas cuya FechaEvento_dt <= cutoff
#    - Aplica mismas columnas y orden
# -------------------------------------------------------
def predict_new(df_new: pd.DataFrame, path_art=os.path.join(ART_DIR, "modelo_calibrado.joblib")):
    art = joblib.load(path_art)
    cutoff = art["cutoff_date"]
    expected_cols = art["all_cols"]
    cal = art["calibrator"]
    thr = art["threshold"]

    assert 'FechaEvento_dt' in df_new.columns, "df_new debe contener FechaEvento_dt"
    # SOLO futuras (jamás vistas)
    df_pred = df_new[df_new['FechaEvento_dt'] > cutoff].copy()
    if df_pred.empty:
        print("No hay filas posteriores al cutoff; nada que predecir.")
        return df_pred.assign(pred_proba=np.nan, pred_label=np.nan)

    # asegurar columnas: crear faltantes como NaN
    Xn = pd.DataFrame(index=df_pred.index)
    for c in expected_cols:
        Xn[c] = df_pred[c] if c in df_pred.columns else np.nan

    proba = cal.predict_proba(Xn)[:,1]
    label = (proba >= thr).astype(int)
    out = df_pred.copy()
    out['pred_proba'] = proba
    out['pred_label'] = label
    keep_cols = ['IdentificadorCliente','FechaEvento_dt'] if 'IdentificadorCliente' in out.columns else ['FechaEvento_dt']
    return out[keep_cols + ['pred_proba','pred_label']]

# -------------------------------------------------------
# 7) ***PRUEBA REAL*** con df_nuevo (estrictamente futuro)
#    Usamos como "df_nuevo" TODO lo posterior al cutoff (holdout).
#    Esto simula el lote que te entregaría negocio después de entrenar.
# -------------------------------------------------------
df_nuevo = clientes.loc[clientes['FechaEvento_dt'] > holdout_cut].copy()
print(f"\ndf_nuevo (futuro real): {df_nuevo.shape}")

# Sanity check: fechas de df_nuevo son posteriores al entrenamiento
assert df_nuevo['FechaEvento_dt'].min() > holdout_cut, "df_nuevo NO es futuro real."

# Predicción sobre df_nuevo (nunca visto en el fit del pipeline)
preds_nuevo = predict_new(df_nuevo)

# Unimos con la verdad para evaluar (porque en esta prueba SÍ la conocemos)
eval_df = preds_nuevo.merge(
    df_nuevo[['FechaEvento_dt', TARGET]],
    left_index=True, right_index=True, how='left'
)
eval_df.rename(columns={TARGET: 'y_true'}, inplace=True)

# Métricas en df_nuevo (futuro nunca visto por el fit)
auc_new = roc_auc_score(eval_df['y_true'], eval_df['pred_proba'])
pr_new  = average_precision_score(eval_df['y_true'], eval_df['pred_proba'])
br_new  = brier_score_loss(eval_df['y_true'], eval_df['pred_proba'])
p2, r2, thr2 = precision_recall_curve(eval_df['y_true'], eval_df['pred_proba'])
f12 = 2*p2*r2/(p2+r2+1e-12)
best2 = np.argmax(f12)
thr2_best = (thr2[best2-1] if best2>0 and best2-1 < len(thr2) else artifact['threshold'])

print(f"\n[DF_NUEVO] ROC-AUC={auc_new:.3f} | PR-AUC={pr_new:.3f} | Brier={br_new:.3f}")
print(f"[DF_NUEVO] best-F1={f12[best2]:.3f} @ thr={thr2_best:.3f}")
yhat_new = (eval_df['pred_proba'] >= thr2_best).astype(int)
print(f"[DF_NUEVO] ConfMatrix @thr={thr2_best:.3f}:\n", confusion_matrix(eval_df['y_true'], yhat_new))
print(f"[DF_NUEVO] Report @thr={thr2_best:.3f}:\n", classification_report(eval_df['y_true'], yhat_new, digits=3))

# (Opcional) Tabla por deciles para negocio (calibración en df_nuevo)
bins = pd.qcut(eval_df['pred_proba'], 10, duplicates='drop')
calib = eval_df.groupby(bins, observed=True).agg(
    p_mean=('pred_proba','mean'),
    y_rate=('y_true','mean'),
    n=('y_true','size')
).reset_index().rename(columns={'pred_proba':'bin'})
print("\n[DF_NUEVO] Calibración por deciles (p_mean vs y_rate):")
print(calib.to_string(index=False))
```

```python
# 1) Parsear fechas
# FechaEvento viene en ISO con 'Z' (UTC). La hacemos tz-aware y luego le quitamos la tz.
clientes['FechaEvento_dt'] = (
    pd.to_datetime(clientes['FechaEvento'], errors='coerce', utc=True)
      .dt.tz_convert(None)  # elimina la tz y deja timestamps "naive"
)

# Fechas que vienen como número de Excel (días desde 1899-12-30)
for c in ['FechaVinculacionCliente','FechaPrimerUso','FechaUltimoUso']:
    clientes[c + '_dt'] = pd.to_datetime(
        clientes[c], errors='coerce', origin='1899-12-30', unit='D'
    )

# 2) Comprobar que todas sean tz-naive
print("FechaEvento_dt tz:", getattr(clientes['FechaEvento_dt'].dt, 'tz', None))
print("FechaVinculacionCliente_dt tz:", getattr(clientes['FechaVinculacionCliente_dt'].dt, 'tz', None))

# 3) Diferencias en días/meses (evitando negativos y nulos)
def safe_months(a, b):
    """(a - b) en meses, con NaN→0 y negativos recortados a 0."""
    delta_days = (a - b).dt.days
    delta_days = delta_days.where(delta_days.notna(), 0)
    delta_days = np.where(delta_days < 0, 0, delta_days)
    return delta_days / 30.0

def safe_days(a, b):
    delta_days = (a - b).dt.days
    delta_days = delta_days.where(delta_days.notna(), 0)
    delta_days = np.where(delta_days < 0, 0, delta_days)
    return delta_days

clientes['MesesDesdeVinculacion'] = safe_months(
    clientes['FechaEvento_dt'], clientes['FechaVinculacionCliente_dt']
)
clientes['MesesDesdePrimerUso'] = safe_months(
    clientes['FechaEvento_dt'], clientes['FechaPrimerUso_dt']
)
clientes['DiasDesdeUltimoUso'] = safe_days(
    clientes['FechaEvento_dt'], clientes['FechaUltimoUso_dt']
)

# 4) Verifica resultados
print(clientes[['FechaEvento_dt','FechaVinculacionCliente_dt','FechaPrimerUso_dt','FechaUltimoUso_dt']].head())
print(clientes[['MesesDesdeVinculacion','MesesDesdePrimerUso','DiasDesdeUltimoUso']].describe())
```

```python
TARGET = 'PerdidaCartera'

# ---------- 1) USABILIDAD CUPO ----------
# Convertir, cap, flags e imputación
clientes['UsabilidadCupo'] = pd.to_numeric(clientes['UsabilidadCupo'], errors='coerce')
clientes['Flag_Usab_NaN']   = clientes['UsabilidadCupo'].isna().astype(int)
# Cap extremo por reglas de negocio: ratio esperado [0,1], toleramos hasta 2 por seguridad
clientes['Flag_Usab_Outlier'] = ((clientes['UsabilidadCupo'] < 0) | (clientes['UsabilidadCupo'] > 2)).astype(int)
clientes.loc[clientes['UsabilidadCupo'] < 0, 'UsabilidadCupo'] = np.nan
clientes.loc[clientes['UsabilidadCupo'] > 2, 'UsabilidadCupo'] = np.nan
# Imputar con mediana
clientes['UsabilidadCupo'] = clientes['UsabilidadCupo'].fillna(clientes['UsabilidadCupo'].median())

print("\n=== Auditoría UsabilidadCupo ===")
print(clientes[['UsabilidadCupo','Flag_Usab_NaN','Flag_Usab_Outlier']].describe(include='all').to_string())

# ---------- 2) FLAGS DE FECHAS (LEAKAGE) ----------
# Ultimo uso posterior al evento -> flag
clientes['Flag_UltimoUsoPosterior'] = (clientes['FechaUltimoUso_dt'] > clientes['FechaEvento_dt']).fillna(False).astype(int)

print("\n=== Flag_UltimoUsoPosterior (cuenta) ===")
print(clientes['Flag_UltimoUsoPosterior'].value_counts(dropna=False).to_string())

# ---------- 3) NUMÉRICAS CON NULOS: imputar 0 + flag ----------
num_cols_candidates = [
    'DiasMaximosMoraCreditosGenerados',
    'NumeroCreditosGPrevius','NumeroCreditosGCanalFPrevius','NumeroCreditosGCanalVPrevius',
    'NumeroCreditosGEstadoActivosPrevius','NumeroCreditosGEstadoPagadosPrevius',
    'NumeroCreditosLPrevius','NumeroCreditosLEstadoActivosPrevius','NumeroCreditosLEstadoPagadosPrevius',
    'TotalPagosEfectuadosGlobalmentePrevius','TotalPagosEfectuadosLocalmentePrevius',
    'NumeroIntentosFallidos','CupoAprobado','ScoreCrediticio','Edad',
    'MesesDesdeVinculacion','MesesDesdePrimerUso','DiasDesdeUltimoUso'
]

# Crear flags y luego imputar (solo donde tiene NaN)
for c in num_cols_candidates:
    if c in clientes.columns:
        clientes[f'Flag_{c}_NaN'] = clientes[c].isna().astype(int)
        # Para conteos históricos y pagos: NaN ~ sin historial -> 0
        if c.startswith('NumeroCreditos') or c.startswith('TotalPagos'):
            clientes[c] = clientes[c].fillna(0)
        elif c in ['DiasMaximosMoraCreditosGenerados','Edad','NumeroIntentosFallidos']:
            clientes[c] = clientes[c].fillna(0)
        elif c in ['CupoAprobado','ScoreCrediticio']:
            # no imputamos aquí todavía (ver manejo de score/cupo abajo)
            pass

# ---------- 4) SCORE & CUPO ----------
# Score: 0 => sin información
clientes['ScoreSinInfo'] = (clientes['ScoreCrediticio'] == 0).astype(int)
# No imputar el 0 (porque es señal), pero si hay NaN residual, usar mediana de los >0
if clientes['ScoreCrediticio'].isna().any():
    med_pos = clientes.loc[clientes['ScoreCrediticio']>0, 'ScoreCrediticio'].median()
    clientes['ScoreCrediticio'] = clientes['ScoreCrediticio'].fillna(med_pos)

# Cupo: crear versión log
clientes['log_CupoAprobado'] = np.log1p(clientes['CupoAprobado'])
# Si hay NaN residual (0.4%), imputar con mediana log
if clientes['log_CupoAprobado'].isna().any():
    clientes['log_CupoAprobado'] = clientes['log_CupoAprobado'].fillna(clientes['log_CupoAprobado'].median())

print("\n=== Auditoría Score/Cupo ===")
print(clientes[['ScoreCrediticio','ScoreSinInfo','CupoAprobado','log_CupoAprobado']].describe().to_string())

# ---------- 5) CATEGÓRICAS: completar nulos ----------
for c in ['CategoriaPrincipalCredito','UsoAppWeb','Genero','TipoMunicipioEntregaTC','CanalMunicipioEntregaTC']:
    if c in clientes.columns:
        clientes[c] = clientes[c].fillna('Desconocido')

# ---------- 6) CONTROL DE RANGOS / SANITY CHECKS ----------
# Edad plausible
clientes['Flag_Edad_Out'] = (~clientes['Edad'].between(18, 100, inclusive='both')).fillna(False).astype(int)
clientes.loc[clientes['Flag_Edad_Out']==1, 'Edad'] = np.nan
clientes['Edad'] = clientes['Edad'].fillna(clientes['Edad'].median())

# Nunca uses identificadores y fechas crudas como features
drop_from_features = [
    'IdentificadorCliente','FechaEvento','FechaVinculacionCliente','FechaPrimerUso','FechaUltimoUso',
    'FechaEvento_dt','FechaVinculacionCliente_dt','FechaPrimerUso_dt','FechaUltimoUso_dt','FechaPrimerUso_corr',
    'CodigoAlmacenEntregaTC','CodigoAlmacenEntregaTC_str','AlmacenTop20',
    'CodigoMunicipioEntregaTC','MunicipioCat','MunicipioTop20','MesCompra',
    'DiasMora'  # fuga
]
# Asegúrate de crear o tener estas dos si las usas:
if 'Flag_PrimerUsoTemu' not in clientes.columns and 'NumeroCreditosGPrevius' in clientes.columns:
    clientes['Flag_PrimerUsoTemu'] = (clientes['NumeroCreditosGPrevius']==0).astype(int)

# ---------- 7) LISTA FINAL DE FEATURES (baseline) ----------
num_final = [
    'UsabilidadCupo','MesesDesdeVinculacion','MesesDesdePrimerUso','DiasDesdeUltimoUso',
    'NumeroCreditosGPrevius','NumeroCreditosGCanalFPrevius','NumeroCreditosGCanalVPrevius',
    'NumeroCreditosGEstadoActivosPrevius','NumeroCreditosGEstadoPagadosPrevius',
    'NumeroCreditosLPrevius','NumeroCreditosLEstadoActivosPrevius','NumeroCreditosLEstadoPagadosPrevius',
    'TotalPagosEfectuadosGlobalmentePrevius','TotalPagosEfectuadosLocalmentePrevius',
    'NumeroIntentosFallidos','ScoreCrediticio','ScoreSinInfo','CupoAprobado','log_CupoAprobado','Edad',
    'Flag_Usab_NaN','Flag_Usab_Outlier','Flag_PrimerUsoAntesVinc','Flag_UltimoUsoPosterior','Flag_Edad_Out'
] + [f'Flag_{c}_NaN' for c in num_cols_candidates if f'Flag_{c}_NaN' in clientes.columns]

cat_final = [
    'Genero','TipoMunicipioEntregaTC','CanalMunicipioEntregaTC','UsoAppWeb','CategoriaPrincipalCredito',
    'Flag_PrimerUsoTemu'  # binaria, si prefieres tratarla como numérica puedes dejarla en num_final
]

# Quitar columnas ausentes o a excluir
num_final = [c for c in num_final if c in clientes.columns and c not in drop_from_features]
cat_final = [c for c in cat_final if c in clientes.columns and c not in drop_from_features]

X = clientes[num_final + cat_final].copy()
y = clientes[TARGET].copy()

print("\n=== Resumen columnas finales ===")
print("Numéricas (", len(num_final), "):", num_final[:12], "...")
print("Categóricas (", len(cat_final), "):", cat_final)

# ---------- 8) AUDITORÍA POST-LIMPIEZA ----------
# Correlación numérica con el target
print("\n=== Correlación numéricas (post-limpieza) con PerdidaCartera ===")
print(X[num_final].corrwith(y).sort_values(ascending=False).to_string())

# Tablas cruzadas para categóricas
for c in cat_final:
    if X[c].nunique() <= 40:
        print(f"\n=== {c} vs {TARGET} ===")
        print(pd.crosstab(X[c], y, normalize='index').round(3).to_string())

# Balance de clases (importante para métricas a usar)
print("\n=== Balance de clases ===")
print(y.value_counts(normalize=True).rename({0:'% no pérdida',1:'% pérdida'}).round(3).to_string())
```
